# 2. OpenCV

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Import OpenCV

```python
import cv2
```

Get the image with the imread

```python
img=cv2.imread('Images/rock.jpg')
```

### Always check the type

```python
type(img)
```

#### Get the Image shape

```python
img.shape
```

#### Let's see the image with imshow

```python
plt.imshow(img)
```

Until now we were working with Matplotlib and RGB

OpenCV is reading the channel as BGR

We will convert OpenCV to the Channels of the photo

```python
img_fix=cv2.cvtColor(img,
                    cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fix)
```

Scale it to Gray and check the shape

```python
img_gray=cv2.imread('Images/rock.jpg',
                   cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

### Resize the Image

To resize the image we change the Height Width --> Width Height

```python
img_new=cv2.resize(img_fix,
                  (1000,400))
```

```python
plt.imshow(img_new)
```

### Resize with Ratio

```python
width_ratio=0.5
height_ratio=0.5
```

```python
img2=cv2.resize(img_fix,
               (0,0),
               img_fix,
               width_ratio,
               height_ratio)
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

### Flip on Horizontal Axis

```python
img_3=cv2.flip(img_fix,0)

plt.imshow(img_3)
```

### Flip on Vertical Axis

```python
img_3=cv2.flip(img_fix,1)

plt.imshow(img_3)
```

### Flip on Horizontal and Vertical Axis

```python
img_3=cv2.flip(img_fix,-1)

plt.imshow(img_3)
```

### Change the size of our Canvas

```python
last_img=plt.figure(figsize=(10,7))
ilp=last_img.add_subplot(111)
ilp.imshow(img_fix)
```
